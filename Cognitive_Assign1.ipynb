```python
#Assign 6.1
#Assingment 6.1
def add_odd(n):
  sum = 0
  for i in range(1,int(n)+1):
    if (i%2!=0):
      sum+=i
  return sum

n = input("Give n ")
print("Sum of odd numbers from 1 to",n," = ",add_odd(n))

#Assign 6.2
def check_prime(n):
  if n==1:
    return False
  elif n==2:
    return True;
  else:
    for x in range(2,n):
        if(n % x==0):
            return False
    return True

n = input("Give n ")
sum = 0
for i in range(1,int(n)+1):
  if (check_prime(i)):
    sum+=i
print("Sum of prime numbers from 1 to",n," = ",sum)
```
